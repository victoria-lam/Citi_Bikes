```python
# Dependencies
import pandas as pd
import numpy as np
import glob
```

```python
# Read csv files from folder
path = r'C:\Users\small\Desktop\Bootcamp Stuff\Citi_Bikes\Data'
filenames = glob.glob(path + "/*.csv")
frame = pd.DataFrame()

# Create list to store dfs
dfs = []
for file in filenames:
    df = pd.read_csv(file, index_col=None, header=0)
    dfs.append(df)

# Concat dfs 
frame = pd.concat(dfs)
```

```python
frame.head()
```

```python
start = frame.groupby(["start station name", "start station latitude", "start station longitude"]).size().reset_index(name="rides")
start.head()
```

```python
end = frame.groupby(["end station name", "end station latitude", "end station longitude"]).size().reset_index(name="rides")
end.head()
```

```python
# Save dfs as csvs 
start.to_csv("Cleaned data/start_stations.csv")
```

```python
end.to_csv("Cleaned data/end_stations.csv")
```

```python
# Get bike id data 
bikeid = frame.groupby(["bikeid"]).size().reset_index(name="count")
bikeid.head()
```

```python
bikeid.to_csv("Cleaned data/bikeid.csv")
```

```python
# Get birth year and time duration 
rider_time = frame.loc[:,["birth year", "tripduration"]].dropna()
rider_time = rider_time.groupby(["birth year"]).sum()
rider_time.head()
```

```python
rider_count = frame.groupby(["birth year"]).size().reset_index(name="count")
rider_count.head()
```

```python
rider_time.to_csv("Cleaned data/rider_time.csv")
```

```python
rider_count.to_csv("Cleaned data/rider_count.csv")
```
